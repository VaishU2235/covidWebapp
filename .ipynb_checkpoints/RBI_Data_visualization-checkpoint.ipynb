# State-wise Average Daily Wage Rates in Rural India (Men - Non-Agricultural Labourers)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('rbi_wages.csv')
df.head()
```

```python
df.dtypes
```

```python
df.isnull
```

```python
df.drop(11,inplace=True)
```

```python
lst=df.columns
ll=lst.tolist()
type(ll)
ll.pop(0)
ll
```

```python
df
```

```python
df['Average']=(df['2014-15']+df['2015-16']+df['2016-17']+ df['2017-18']+df['2018-19']+df['2019-20'])/6
df
```

```python
df['total_increase']=((df['2019-20']-df['2018-19'])+(df['2018-19']-df['2017-18'])+(df['2017-18']-df['2016-17'])+(df['2016-17']-df['2015-16'])+(df['2015-16']-df['2014-15']))
df['average_increase']=(df['total_increase'])/5
```

```python
df
```

```python
plt.figure(figsize=(15, 10))
plt.xticks(rotation=70)
plt.suptitle('average increase in wages VS states/union territory')
plt.bar(df['State/Union Territory'],df['average_increase'],color=['red'])
```

```python
plt.figure(figsize=(15, 10))
plt.xticks(rotation=45)
plt.suptitle('average wages over the years VS states/union territory')
plt.bar(df['State/Union Territory'],df['Average'],color=['blue'])
```
