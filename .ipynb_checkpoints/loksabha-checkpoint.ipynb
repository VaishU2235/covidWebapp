```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Loksabha_1962-2019 .csv')
df.head()
```

```python
df.shape
```

```python
df[df['Turnout'] == '-']
```

```python
df.replace('-',value=np.nan,inplace=True)
```

```python
df[df['party'] == '-']
```

```python
df[df['Turnout'] == "Margin"]
```

```python
df.drop(3161,axis=0,inplace=True)
```

```python
df['Turnout'] = df['Turnout'].astype(str)

df["Turnout"]=df['Turnout'].map(lambda name: name.split(' ')[0])
df['Turnout'] = df['Turnout'].astype(float)

df['Turnout'].replace(0,value=np.nan,inplace=True)
```

```python
df['margin%']=df['margin%'].str.replace('%', '').astype(float)
df['margin%'].replace(0,value=np.nan,inplace=True)
```

```python
df['electors']=df['electors'].str.replace(',', '').astype(float)
df['electors'].replace(0,value=np.nan,inplace=True)
```

```python
df[df['votes'] == "RU"]
```

```python
df.replace('RU',value=np.nan,inplace=True)
```

```python
df['votes']=df['votes'].str.replace(',', '').astype(float)
df['votes'].replace(0,value=np.nan,inplace=True)
```

```python
df['year']=df['year'].astype(int)
```

```python
df.rename(columns={'Turnout':'turnout'},inplace=True)
```

```python
df.head()
```

```python
df.to_csv("Loksabha_election_data_1962-2019",index=False)
```
