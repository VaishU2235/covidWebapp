```python
import pandas as pd
```

```python
url='https://api.covid19india.org/csv/latest/states.csv'
df=pd.read_csv(url)
df.head()
```

```python
df.info()
```

```python
df['Date']=pd.to_datetime(df['Date'],infer_datetime_format=True)
```

```python
df.info()
```

```python
df['month']=df['Date'].dt.to_period('M')
df.tail()
```

```python
df.head()
```

```python
df.groupby('month')
```

```python
ls=df["month"].unique()
ls
```

```python
l=ls[-2]
l
```

```python
d=df.groupby('month').get_group(l).groupby("State").get_group('Kerala')
```

```python
d.reset_index(inplace=True)
d
```
